```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load the dataset

```python
dataset = pd.read_csv('Medical_Price_Dataset.csv')
```

Extract the features (X) and target variable (y)

```python
X = dataset[['age', 'bmi', 'children']]
y = dataset['charges']
```

Add a column of ones to X for the intercept term

```python
X['intercept'] = 1
```

Convert X and y to NumPy arrays

```python
X = X.values
y = y.values
```

Calculate the optimal weights using the normal equation

```python
weights = np.linalg.inv(X.T @ X) @ X.T @ y
```

Predict the target variable using the learned weights

```python
y_pred = X @ weights
```

Calculate the mean squared error

```python
mse = np.mean((y_pred - y) ** 2)
```

Print the learned weights and mean squared error

```python
print("Learned Weights:", weights)
print("Mean Squared Error:", mse)
```

Plot the predicted values against the actual values

```python
plt.scatter(y, y_pred)
plt.plot([np.min(y), np.max(y)], [np.min(y), np.max(y)], 'r--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
